```python
import requests


def len_joke():
	"""

	"""
	joke = get_joke()
	return len(joke)


def get_joke():
	url = 'http://api.icndb.com/jokes/random'

	response = requests.get(url)

	if response.status_code == 200:
		joke = response.json()['value']['joke']
	else:
		joke = 'No jokes'

	return joke
```

```python
import datetime as dt
import os
import pandas_datareader as pdr
import yfinance as yf

yf.pdr_override()


def main():
	save_sp500_tickers()
	get_data_from_yahoo(reload_sp500=False)


def save_sp500_tickers():
	"""A function to save sp500 data as pickle"""
	# TODO modify to change to csv
	resp = requests.get('http://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
	soup = bs.BeautifulSoup(resp.text, 'lxml')
	table = soup.find('table', {'class': 'wikitable sortable'})
	tickers = []
	for row in table.findAll('tr')[1:]:
		ticker = row.findAll('td')[0].text
		ticker = ticker[:-1]
		tickers.append(ticker)

	with open("sp500tickers.pickle", "wb") as f:  # TODO remove pickle swap csv
		pickle.dump(tickers, f)
	return tickers


def get_data_from_yahoo(reload_sp500=False):
	"""
	parsing sp500 pickel into csv
	:param reload_sp500:
	:return:
	"""
	if reload_sp500:
		tickers = save_sp500_tickers()
	else:
		with open("sp500tickers.pickle", "rb") as f:
			tickers = pickle.load(f)
	if not os.path.exists('stock_dfs'):
		os.makedirs('stock_dfs')

	start = dt.datetime(1900, 1, 1)
	end = dt.datetime.now()

	for ticker in tickers:
		if not os.path.exists('{}/stock_dfs/{}.csv'.format(CSV_SAVE_PATH, ticker)):
			df = yf.download(ticker.replace('.', '-'), start, end, interval='1d')
			df.reset_index(inplace=True)
			df.set_index("Date", inplace=True)
			df.to_csv('{}/stock_dfs/{}.csv'.format(CSV_SAVE_PATH, ticker))
			print('Ticker {} updated'.format(ticker))
		else:
			print('Already have {}'.format(ticker))


if __name__ == '__main__':
	CSV_SAVE_PATH = '~/sbox/data_dirty/ya_assets/'
	main()
```

```python
# ALPHA VANTAGE2
import csv
import requests

# replace the "demo" apikey below with your own key from https://www.alphavantage.co/support/#api-key
CSV_URL = 'https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY_EXTENDED&symbol=IBM&interval=15min&slice=year1month1&apikey=demo'

with requests.Session() as s:
	download = s.get(CSV_URL)
	decoded_content = download.content.decode('utf-8')
	cr = csv.reader(decoded_content.splitlines(), delimiter=',')
	my_list = list(cr)
	for row in my_list:
		print(row)
```
